# 1 美国各州人口密度分析

```python
import numpy as np
import pandas as pd
```

## 1.1 首先导入文件，并查看数据样本

```python
state_abbrevs = pd.read_csv('data/state-abbrevs.csv')
state_abbrevs.head()
```

```python
state_areas = pd.read_csv('data/state-areas.csv')
state_areas.head()
```

```python
state_population = pd.read_csv('data/state-population.csv')
state_population.head()
```

```python
display(state_abbrevs.head(),state_areas.head(),state_population.head())
```

## 1.2 合并state_abbrevs和state_populations数据集，分别根据abbreviation和state/region列进行合并，得到一个新的数据集state_info

```python
state_info = pd.merge(left = state_population, right = state_abbrevs, how = 'outer', left_on = 'state/region', right_on = 'abbreviation')
```

```python
state_info
```

## 1.3 查看存在缺失数据的列

```python
state_abbrevs.loc[:, 'abbreviation'].unique() # 查看state_abbrevs数据集中的abbreviation列中有哪些唯一值
```

```python
state_population.loc[:, 'state/region'].unique() # 查看state_population数据集中的state/region列中有哪些唯一值   
```

```python
# PR 和 USA在state_population数据集中的state/region列中，但不在state_abbrevs数据集中的abbreviation列中
set(state_population.loc[:, 'state/region']) - set(state_abbrevs.loc[:, 'abbreviation']) # 查看state_population数据集中的state/region列中有哪些唯一值不在state_abbrevs数据集中的abbreviation列中
```

```python
state_info.loc[:, 'state/region'].unique() # 查看state_info数据集中的state/region列中有哪些唯一值
```

```python
state_info.isnull().any() # 查看state_info数据集是否有缺失值
```

## 1.4 根据数据是否缺失情况显示数据，如果缺失为True，则显示

```python
state_info.isnull().any(axis=1) # 查看state_info数据集是否有缺失值
```

```python
state_info.loc[state_info.isnull().any(axis=1)] # 根据数据是否缺失情况显示数据，如果缺失为True，则显示 
```

## 1.5 找到有哪些state/region使得state的值为Nan，使用unique()方法查看非重复值

```python
state_info.loc[state_info.isnull().any(axis=1)].loc[:, 'state/region'].unique() # 找到有哪些state/region使得state的值为Nan，使用unique()方法查看非重复值
```

## 1.6 为找到的这些state/region的state项补上正确的值，从而去掉state这一列的所有Nan

```python
# 经过分析，只有PR和USA这两个state/region的state值为Nan，因此我们需要为这两个state/region的state项补上正确的值，从而去掉state这一列的所有Nan
# USA -> USA
# PR -> Puerto Rico
```

```python
# 面积表中存在PR对应的州名为Puerto Rico
set(state_areas.loc[:, 'state'])-set(state_abbrevs.loc[:, 'state']) # 查看state_areas数据集中的state列中有哪些唯一值不在state_abbrevs数据集中的state列中
```

```python
# 将state_info数据集中的PR对应的state项补上Puerto Rico
state_info.loc[state_info['state/region'] == 'PR', 'state'] = 'Puerto Rico'
```

```python
state_info.loc[state_info.isnull().any(axis=1)] # 根据数据是否缺失情况显示数据，如果缺失为True，则显示 
```

```python
# 由于USA是全美国的数据，直接删掉
state_info.drop(state_info[state_info['state/region'] == 'USA'].index, inplace=True)
```

```python
state_info.loc[:, 'state/region'].unique() # 查看state_info数据集中的state/region列中有哪些唯一值
```

```python
state_info.isnull().any() # 查看state_info数据集是否有缺失值
```

```python
state_info
```

```python
# 由于abbreviation列有空值，将这一列删除
state_info.drop('abbreviation', axis=1, inplace=True)
```

```python
state_info
```

## 1.7继续寻找存在缺失数据的列

```python
# 找到population列中存在缺失值的行
state_info.loc[state_info.population.isnull()]
```

```python
# 由于PR部分年份没有数据，将这部分删掉
state_info['population'].isnull()
```

```python
pop_index = state_info.loc[state_info['population'].isnull()].index
pop_index
```

```python
state_info.drop(pop_index, axis=0, inplace=True)
```

```python
state_info['population'].isnull().any() # 查看population列中是否还有缺失值
```

```python
# 法二
state_info = state_info.dropna()
```

## 1.8 合并各州面积数据

```python
total = pd.merge(left = state_info, right = state_areas, how = 'outer', left_on = 'state', right_on = 'state')
```

```python
total
```

## 1.9 找出2010年的全年总人口数据，df.query（查询语句）

```python
pop_2010 = total.query('year == 2010 & ages == "total"') # 找出2010年的全年人口数据
pop_2010
```

```python
total['year'] == 2010
```

```python
total['ages'] == 'total'
```

```python
con = (total['year'] == 2010) & (total['ages'] == 'total') # 找出2010年的全年人口数据
```

```python
total.loc[con] # 根据条件con显示数据
```

## 1.10 对查询结果进行处理，以state列作为新的行索引：set_index()方法

```python
total.set_index('state', inplace=True)
```

```python
total
```

## 1.11 计算各州人口密度，使用人口总数除以面积，得到一个新的列density

```python
total['density'] = total['population'] / total['area (sq. mi)'] # 计算各州人口密度，使用人口总数除以面积，得到一个新的列density
```

```python
total
```

## 1.12 对查询结果进行排序

```python
total.sort_values(by='density', ascending=True, inplace=True)
```

```python
total
```
