# 1 Matplotlib

## 1.1 获取画布

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.figure(figsize=(8,2))
```

```python
# 创建一个画布，画布是不呈现图像的，画板是成像的，画布只是用来管理画板，画布的比例会影响画板的比例，画布的大小会影响图像的大小。
figure = plt.figure(figsize=(6,4)) 
```

## 1.2 获取画板

```python
# 在画布上创建一个画板
# 111 可以拆成三个数字：
# 第一个 1：画布分成 1 行
# 第二个 1：画布分成 1 列
# 第三个 1：选择第一个子图
# 所以它表示：创建一个占满整个画布的子图。
# 使用 plt.subplot() 时，画板不需要显式的绑定画布，Matplotlib 会把画板放到离当前画板最近的画布上
axes = plt.subplot(111)
axes
```

```python
# 最推荐的方式是使用 plt.subplots()，它会同时创建画布和画板，并返回一个元组，包含画布和画板对象。
fig, axes = plt.subplots(figsize=(8,6)) 
axes
```

## 1.3 就近原则

plt.的所有操作都是在离当前画板最近的画布上进行的。

所有的画板，都会默认配置在离当前画板最近的画布上。<br>
如果当前画板没有绑定画布，Matplotlib 会自动创建一个画布，并把画板绑定到这个画布上。

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
```

```python
figure = plt.figure(figsize=(4,4))
axes1 = plt.subplot(221) # 创建一个 2x2 网格中的第一个子图
axes2 = plt.subplot(224) # 创建一个 2x2 网格中的第四个子图
# 绘制正弦曲线（面向过程风格）
plt.plot(x, y)# 这里也体现了就近原则
    
# 指定画板绘制图像（面向对象风格）
axes1.plot(x, y)
axes1.plot(x, 2*y)
```

## 1.4 网格线

面向过程 --> plt.<br>
面向对象 --> axes.

```python
fig, axes = plt.subplots(figsize=(6, 4))
plt.grid() # 面向过程
```

```python
figure = plt.figure(figsize=(8, 4))
axes1 = plt.subplot(121) # 创建一个 1x2 网格中的第一个子图
axes2 = plt.subplot(122) # 创建一个 1x2 网格中的第二个子图
plt.grid() # 就近原则，当前最近画板是 axes2，所以网格线绘制在 axes2 上
axes1.grid(color ='red', linestyle='--') # 指定画板绘制网格线
```

## 1.5 坐标轴界限

```python
fig, axes = plt.subplots(figsize=(6, 4))
axes.plot(x, y)
# 默认情况下，Matplotlib 会根据数据的范围自动设置坐标轴的界限。

# 1 使用 set_xlim() 和 set_ylim() 方法可以手动设置坐标轴的界限。
axes.set_xlim(0, 2*np.pi+3)# x 轴的界限设置为 0 到 2π+3
axes.set_ylim(-2, 2)# y 轴的界限设置为 -2 到 2
```

```python
# 2 也可以使用 plt.xlim() 和 plt.ylim() 方法来设置坐标轴的界限，这种方式是面向过程的。
plt.plot(x, y)
plt.xlim(0, 2*np.pi+3)# x 轴的界限设置为 0 到 2π+3
plt.ylim(-2, 2)# y 轴的界限设置为 -2
```

```python
# 3 使用axis()方法可以同时设置 x 轴和 y 轴的界限，传入一个列表，列表中包含四个元素，分别表示 x 轴的最小值、最大值，y 轴的最小值、最大值。
plt.plot(x, y)
axes.axis((0, 2*np.pi+3, -2, 2))# x 轴的界限设置为 0 到 2π+3，y 轴的界限设置为 -2 到 2
```

## 1.6 设置刻度

axes.set_xticks() 和 axes.set_yticks() 方法可以设置刻度的位置和标签。<br>
- 位置：传入一个列表，列表中包含刻度的位置。<br>
- 标签：传入一个列表，列表中包含刻度的标签。<br>
- 使用 plt.xticks() 和 plt.yticks() 方法也可以设置刻度的位置和标签，这种方式是面向过程的。

```python
fig, axes = plt.subplots(figsize=(6, 4))
axes.plot(x, y)
# 设置 x 为[0, 2*pi]，y 为[-1, 1]，并且设置 x 轴的刻度为[0, pi/2, pi, 3*pi/2, 2*pi]，y 轴的刻度为[-1, -0.5, 0, 0.5, 1]
_=axes.set_xticks(ticks=[0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
_=axes.set_yticks(ticks=[-1, 0, 1], labels=['min', '0', 'max'])
```

### 1.7 坐标轴标签

坐标轴标签是对整个 x 轴和 y 轴含义的说明，可使用 set_xlabel() 和 set_ylabel()：
- 也可以使用 plt.xlabel() 和 plt.ylabel() 方法来设置坐标轴标签，这种方式是面向过程的。

```python
fig, axes = plt.subplots(figsize=(6, 4))
axes.plot(x, y)
axes.set_xlabel('x_label')# 设置 x 轴标签
axes.set_ylabel('y_label')# 设置 y 轴标签
```

## 1.10 图例

### 1.10.1 两种传参方式

- 分别在plot函数中传入label参数，再调用plt.legend()方法显示图例。<br>
- 直接在plt.legend()方法中传入label参数显示图例。

```python
import matplotlib.pyplot as plt
# 方式一：在 plot() 中传入 label参数，然后调用 plt.legend() 方法显示图例。
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 8, 27, 64, 125]

fig, axes = plt.subplots(figsize=(6, 4))
axes.plot(x, y1, label='y=x^2')# 在 plot() 中传入 label参数
axes.plot(x, y2, label='y=x^3')#
plt.legend()# 调用 plt.legend() 方法显示图例
plt.show()# plt.show() 用于显示当前绘制的图形，再调用 plt.show() 之后，将会重新生成画布和画板
```

```python
# 方式二：直接在 plt.legend() 中传入标签
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 8, 27, 64, 125]
plt.plot(x, y1,c='r')# 绘制 y=x^2 曲线，颜色为红色
plt.plot(x, y2)
plt.legend(['y=x^2', 'y=x^3'])# 按绘图顺序设置图例，直接在 plt.legend() 中传入标签
plt.show()# plt.show() 用于显示当前绘制的图形，再调用 plt.show() 之后，将会重新生成画布和画板
```

```python
# 面向对象风格
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 8, 27, 64, 125]
fig, axes = plt.subplots(figsize=(6, 4))
axes.plot(x, y1, label='y=x^2')# 在 plot() 中传入 label参数
axes.plot(x, y2, label='y=x^3')
axes.legend()# 调用 axes.legend() 方法显示图例
plt.show()# plt.show() 用于显示当前绘制的图形，再调用 plt.show() 之后，将会重新生成画布和画板
```

## 1.10.2 loc参数

- 使用 legend() 的 loc 参数修改图例位置，常用的取值有：
  - 'best'：自动选择最佳位置（默认值）<br>
  - 'upper right'：右上角<br>
  - 'upper left'：左上角<br>
  - 'lower left'：左下角<br>
  - 'lower right'：右下角<br>
  - 'right'：右侧<br>
  - 'center left'：左侧中间<br>
  - 'center right'：右侧中间<br>
  - 'lower center'：下方中间<br>
  - 'upper center'：上方中间<br>
  - 'center'：中心

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 8, 27, 64, 125]
plt.plot(x, y1, label='y=x^2', c='r')# 绘制 y=x^2 曲线，颜色为红色
plt.plot(x, y2, label='y=x^3', c='b')# 绘制 y=x^3 曲线，颜色为蓝色
plt.legend(loc='upper right')# 设置图例位置为右上角
plt.show()# 显示图形
```

loc参数还可以传入一个二元组，表示图例的坐标位置
- [0, 1]：左上角<br>
- [1, 0]：右下角<br>
- [0.5, 0.5]：中心<br>
- [0, 0]：左下角<br>
- [1, 1]：右上角<br>

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
plt.plot(x, y1, label='y=x^2', c='r')# 绘制 y=x^2 曲线，颜色为红色
plt.legend(loc = [0.5, 0.5])# 设置图例位置为坐标(0.5, 0.5)
```

### 1.10.3 ncol参数
- 使用 legend() 的 ncol 参数修改图例的列数，ncol=3

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
plt.plot(x, [1,2,3,4,5], label='y=x') # 绘制 y=x 曲线
plt.plot(x, [1,4,9,16,25], label='y=x^2') # 绘制 y=x^2 曲线
plt.plot(x, [1,8,27,64,125], label='y=x^3') # 绘制 y=x^3 曲线
plt.legend(ncol=3) # 设置图例为三列
plt.show()# 显示图形
```

### 1.10.4 linestyle、color、marker参数
- 使用 legend() 的 linestyle、color、marker 参数修改图例的线型、颜色和数据点样式

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, label='y=x^2', color='r', linestyle='--', marker='o')# 绘制 y=x^2 曲线，颜色为红色，线型为虚线，数据点为圆形
plt.legend()# 显示图例
plt.show()# 显示图形
```

## 1.11 图片保存

```python
#lambda 表达式是 Python 中创建“匿名函数”的简洁写法，适合定义简单、只使用一次的函数。
# lambda 参数: 表达式
import numpy as np
x = np.linspace(-1, 1, 100)
f = lambda x: (1-x**2)**0.5
y = f(x)
```

```python
figure = plt.figure(figsize=(4,4),facecolor='green')# 创建一个画布，大小为 4x4 英寸，背景颜色为绿色
axes = plt.subplot(111)
axes.plot(x, y,c='r')
axes.plot(x, -y,c='r')
figure.savefig('circle.png', dpi=300, bbox_inches='tight')# 保存图像，dpi=300 表示分辨率为 300，bbox_inches='tight' 表示去掉多余的空白
```

# 2. 添加注释

## 2.1 坐标轴区域的文字注释

- text() 用来在图中的指定位置添加文字说明
- plt.text() 的文字默认显示在“坐标轴区域”内，也就是图中有坐标刻度、曲线的那块绘图区。

```python
import matplotlib.pyplot as plt
figure = plt.figure(figsize=(10, 4),facecolor='green')# 创建一个画布，大小为 10x4 英寸，背景颜色为绿色
axes1 = plt.subplot(121)
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, label='y=x^2', c='r', linestyle='--', marker='o')# 绘制 y=x^2 曲线，颜色为红色，线型为虚线，数据点为圆形
plt.legend()# 显示图例
# x，y 表示文字注释的位置，x表示x轴坐标，y表示y轴坐标，s 表示文字内容
plt.text(x = 3, y = 12.5, s='dou fu',fontdict={'size': 12, 'color': 'blue'} )

axes2 = plt.subplot(122)
plt.text(x=0.5, y=0.5, s='Hello World')
```

## 2.2 画布上的文字注释

figtext() 用于在整张画布（Figure）的指定位置添加文字

```python
import matplotlib.pyplot as plt

figure = plt.figure(figsize=(10, 4),facecolor='green')
axes1 = plt.subplot(311)
axes2 = plt.subplot(313)

#添加画布注释，x，y 表示文字注释的位置，表示的是相对位置
plt.figtext(x=0.5, y=0.5, s='Hello World',color='blue', fontsize=20)
```

## 2.3 添加画布标题

给整张画布添加标题，使用 fig.suptitle()

```python
# x,y 表示标题位置，使用相对坐标，x=0.5 表示水平居中，y=0.95 表示距离画布顶部 5% 的位置
import matplotlib.pyplot as plt
figure = plt.figure(figsize=(10, 4),facecolor='green')
axes = plt.subplot(111)
x= [1, 2, 3, 4, 5]
y= [2, 4, 6, 8, 10]
plt.plot(x, y)
plt.suptitle('This is a figure title', fontsize=16, color='red',x = 0.1, y = 0.95)# 设置画布标题，x，y 表示标题位置，x=0.5 表示水平居中，y=0.95 表示距离画布顶部 5% 的位置
```

## 2.4 添加箭头注释

Matplotlib 使用 annotate() 添加带箭头的注释。

常用参数：<br>
xy设置箭头指向的位置<br>
xytext设置注释文字的位置<br>
arrowprops以字典的形式设置箭头样式<br>

如下都是arrowprops可以选择的分格样式：<br>
"->"      # 普通单向箭头<br>
"<->"     # 双向箭头<br>
"-|>"     # 三角形箭头<br>
"fancy"   # 花式箭头<br>

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
figure = plt.figure(figsize=(10, 4))
axes = plt.subplot(111)
plt.plot(x, y)
plt.suptitle('This is a figure title', fontsize=16, color='red',x = 0.5, y = 0.95)# 设置画布标题，x，y 表示标题位置，x=0.5 表示水平居中，y=0.95 表示距离画布顶部 5% 的位置
plt.legend(['y=x^2'])# 设置图例
plt.annotate('max', xy=(5, 25), xytext=(5, 20), arrowprops=dict(arrowstyle='->', color='green'))# 添加注释，xy 表示箭头指向的位置，xytext 表示文字注释的位置，arrowprops 表示箭头的属性
plt.show()
```

### 3.1.2 透明度

### 3.1.3 背景色

### 3.1.4 线型

Matplotlib 使用 linestyle 或简写ls设置线型。

"-"----->实线 <br>
"--"----->虚线 <br>
"-."----->点划线 <br>
":"----->点线 <br>
"None"----->不显示线 <br>

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.sin(x)

plt.plot(x, y, linestyle='--')# 实线
plt.plot(x, 2*y, linestyle='-.')# 点划线
plt.plot(x, 3*y, linestyle=':')# 点线
plt.plot(x, 4*y, 'r--')# 简写
plt.show()
```

### 3.1.5 线宽

Matplotlib 使用 linewidth 或简写 lw 设置线宽。

```python
import matplotlib.pyplot as plt
x = np.linspace(-np.pi, np.pi, 100)
y = np.sin(x)
plt.plot(x, y, linewidth=1)# 线宽为 1
plt.plot(x, 2*y, linewidth=2)# 线宽为 2
plt.plot(x, 3*y, lw=3,color='r',linestyle='--')# 线宽为 3
plt.show()
```

#### 3.1.5.1 不同宽度的破折线

dashes 用于自定义虚线的线段和间隔长度

设置破折号序列各段段宽度

```python
import matplotlib.pyplot as plt
x= np.linspace(-np.pi, np.pi, 100)
y= np.sin(x)
plt.plot(x, y, dashes=[10, 2, 5, 2])# 设置破折号序列为 5 个点的线段和 2 个点的间隔
plt.show()
```

### 3.1.6 点型

Matplotlib 使用 marker 参数设置点型，也就是数据点的样式。</b>
- markersize设置点的大小
- markeredgewidth设置点边缘的宽度
- markeredgecolor设置点边缘的颜色
- markerfacecolor设置点的填充颜色。

常见的点型有：
- "."：点<br>
- ","：像素点<br>
- "o"：圆圈<br>
- "v"：倒三角<br>
- "^"：正三角<br>
- "<"：左三角<br>
- ">"：右三角<br>
- "1"：下箭头<br>
- "2"：上箭头<br>
- "3"：左箭头<br>
- "4"：右箭头<br>

```python
import matplotlib.pyplot as plt
x = np.linspace(-np.pi, np.pi, 10)
y = np.sin(x)
plt.plot(x, y, linestyle='None', marker='o')# 线宽为 1, marker='o' 表示数据点为圆形
plt.plot(x, 2*y, linestyle='None', marker='s')# 线宽为 2, marker='s' 表示数据点为方形
plt.plot(x, 3*y, color='r',linestyle='None',marker='^',markersize=20)# 线宽为 3, 颜色为红色, 线型为虚线, marker='^' 表示数据点为正三角
```

### 3.1.7 多参数连用


颜色、点型、线型，可以把几种参数连用在一起，形成不同的组合，来设置数据点的样式。

```python
import matplotlib.pyplot as plt
x = np.linspace(-np.pi, np.pi, 10)
y = np.sin(x)
plt.plot(x, y, 'bo--')# 颜色为蓝色，点型为圆形，线型为虚线
plt.plot(x, 2*y, 'g^-.')# 颜色为绿色，点型为正三角，线型为点划线
plt.plot(x, 3*y, 'r*:')# 颜色为红色，点型为星形，线型为点线
plt.show()
```
