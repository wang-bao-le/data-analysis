## 1.2 转换时间戳

### 1.2.1 to_datetime()实现批量日期转换

to_datetime()函数可以将字符串、纪元式及混合的日期Series或者日期列表转换为pandas的时间戳类型，返回一个DatetimeIndex对象。<br>
转换的是Series时，返回的是一个Series对象，索引不变，值为时间戳类型。<br>
转换的是列表时，返回的是一个DatetimeIndex对象。<br>

pandas可以识别的日期格式有：
- '%Y-%m-%d' (例如: 2023-01-01)
- '%Y/%m/%d' (例如: 2023/01/01)
- '%Y.%m.%d' (例如: 2023.01.01)

```python
s = pd.Series([
    '2005/11/23',
    '2010.06.24',
    'Jul,31, 2009',
    '2012-09-08',
    'Aug-09-2011',
    'Feb 1 2010',
    '01-14-2012',
    None],name='date')
```

```python
s
```

```python
df = pd.DataFrame(data = s)
df
```

```python
df['value'] = np.random.randint(1,10,8)
df
```

```python
df.sort_values(by = 'date') # 不是按照时间戳排序，而是按照字符串排序
```

```python
pd.to_datetime(df['date'], format='mixed', errors='coerce')  # 将字符串转换为时间戳
```

#### 时间类型的字符串，可以直接使用to_datetime()函数转换为Timestamp类型，转换后的时间戳可以直接进行时间运算。<br>
#### 并且空值会被自动转换为NaT类型，NaT是pandas中时间戳的空值类型。<br>

```python
df['date'] = pd.to_datetime(df['date'], format='mixed', errors='coerce')  # 将字符串转换为时间戳类型
```

```python
df
```

```python
df.sort_values(by='date',ascending=True)
```

解析欧式日期（日-月-年），要用dayfirst=True参数，解析美式日期（月-日-年），要用dayfirst=False参数。<br>

```python
pd.to_datetime('04-01-2023', dayfirst=True)  # 解析欧式日期（日-月-年）
```

```python
pd.to_datetime('04-01-2023', dayfirst=False)  # 解析美式日期（月-日-年）
```

dayfirst并没有那么严苛，如果不能把第一个数字解释为日，则会尝试把第一个数字解释为月。<br>

```python
pd.to_datetime(['2023-01-01', '02-14-2023'], dayfirst=True, format='mixed')  # 将字符串列表转换为时间戳类型
```

```python
# to_datetime()转换单个字符串时，返回的是Timestamp类型，转换多个字符串时，返回的是DatetimeIndex类型。
pd.to_datetime('2023-01-01')  # 单个字符串
```

Timestamp函数仅支持单个字符串的转换，不支持dayfirst、format参数。如果要使用dayfirst、format参数，需要使用to_datetime()函数。<br>

```python
 # 指定格式转换,format参数指定字符串的格式，%Y表示四位数的年份，%m表示两位数的月份，%d表示两位数的日期
pd.to_datetime('2023年01月01日', format='%Y年%m月%d日') 
```

### 1.2.4 无效数据

不可解析时，默认值errors='raise'，会抛出异常。可以通过errors参数指定错误处理方式，errors参数的可选值有：
- 'raise'：抛出异常，默认值
- 'coerce'：将无法解析的值设置为NaT

```python
index = pd.to_datetime(['2023-01-01', 'asdf'], errors='coerce')  # 将字符串列表转换为时间戳类型，无法解析的值会被设置为NaT
```

```python
index[0],index[1]  # 获取第一个时间戳和第二个时间戳
```

```python
data=pd.Series(data = ['2012-01-01','sade', '2013年01月01日',None],name='date')  # 创建一个包含无效数据的Series对象
```

```python
data
```

```python
def transform_date(date_str):
    # 处理无效数据
    if pd.isna(date_str):#pd.isna可以判断是否为缺失值，缺失值包括None和NaN
        return pd.NaT
    # 处理中文日期格式
    if date_str.endswith('年') or date_str.endswith('月') or date_str.endswith('日'):
        return pd.to_datetime(date_str, format='%Y年%m月%d日', errors='coerce')  # 将字符串转换为时间戳类型，无法解析的值会被设置为NaT
    else:# 处理字符串
        obj = pd.to_datetime(date_str, format='mixed', errors='coerce')  # 将字符串转换为时间戳类型，无法解析的值会被设置为NaT
        return obj
```

```python
data.map(transform_date)  # 使用map()方法将transform_date函数应用于Series对象的每个元素
```

## 1.6 重采样

pandas有一个虽然简单但非常强大的功能，可在频率转换时执行重采样，如，将秒数据转换为分钟数据，或者将日数据转换为月数据。<br>

resample()函数是基于时间序列的groupby()函数，每个组都遵循归纳方法。

resample（）可以直接用于dataframegroupby()函数

> 注意：重采样一定要以时间类型做index

```python
aapl
```

```python
aapl.resample('YE')['Open'].mean() # 按年重采样，计算每年的平均值
```

```python
# 每三年平均开盘价
aapl.resample('3YE')['Open'].mean() # 按三年重采样，计算每三年的平均值
```

```python
# 每450的平均开盘价
aapl.resample('450D')['Open'].mean() # 按450天重采样，计算每450天的平均值
```

## 1.7 时间差

Timedelta（）:时间差对象，表示两个时间点之间的差值。可以通过减法运算符（-）计算两个时间戳之间的差值，返回一个Timedelta对象。

```python
now = pd.Timestamp.now()  # 获取当前时间戳
```

```python
now
```

```python
delta = pd.Timedelta(4, unit='D')  # 创建一个表示4天的时间差对象
delta
```

```python
now + delta
```

```python
delta = pd.Timedelta(days=4, hours=3, minutes=30)  # 创建一个表示4天3小时30分钟的时间差对象
delta
```

```python
now + delta
```

以时间差为数据的Series和DataFrame对象支持各种运算，datetime64[ns]序列或Timestamps减法运算生成的是timedelta64[ns]序列。

```python
s = pd.Series(data = pd.date_range(start='2023-01-01', periods=5, freq='D'))  # 创建一个以时间戳为数据的Series对象
```

```python
d = pd.Series(data = [pd.Timedelta(days=i) for i in range(5)])  # 创建一个以时间差为数据的Series对象
```

```python
df = pd.DataFrame(data = {'date': s, 'delta': d})  # 创建一个包含时间戳和时间差的DataFrame对象
df
```

```python
df['new_date'] = df['date'] + df['delta']  # 将时间戳和时间差相加，得到新的时间戳
```
