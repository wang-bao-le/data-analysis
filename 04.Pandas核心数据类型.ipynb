# Pandas 核心数据类型


## 什么是Pandas
- pandas是python语言的一个扩展程序库，用于数据分析。
- pandas提供了高性能、易用的数据结构和数据分析工具，使得数据分析变得更加简单和高效。
- pandas一个强大的分析结构化数据的工具集，基础是NumPy（提供高性能的矩阵运算）
- pandas可以从各种文件格式比如CSV、Excel、SQL数据库等中读取数据
- pandas可以对各种数据进行运算操作，比如归并、再成形、选择、还有数据清洗和数据加工特征

#### pandas的主要数据结构
- Series: 一维标签化数组，可以存储任何数据类型（整数、字符串、浮点数、Python对象等）。每个元素都有一个标签（索引），可以通过标签访问数据。
- DataFrame: 二维标签化数据结构，类似于电子表格或SQL表格。它由行和列组成，每个列可以是不同的数据类型。

## Series
Series是一种类似于字典（dict-like）的一维数组（array-like）对象。它由两部分组成：数据（values）和索引（index）。
- 数据（values）是一个包含任意数据类型的数组（ndarray），而索引（index）是一个与数据对应的标签数组。
- Series的索引可以是整数、字符串或其他类型的标签。默认情况下，Series会自动生成整数索引（从0开始），但你也可以自定义索引。
