## ndarray的级联和切分

### 级联
级联的注意事项：
- 级联的参数是列表：一定要加中括号或者小括号
- 维度必须相同（都是一维/二维）
- 形状相符（在级联的方向上必须长度一致）
- 级联的方向默认是shape这个tuple的第一个值所代表的维度方向
- 可通过axis参数改变级联的方向

np.concatenate(arrays，axis)函数可以将多个数组沿着指定的轴进行连接，返回一个新的数组
参数说明：
- arrays：要连接的数组序列，可以是一个列表或元组，包含了要连接的数组。
- axis：沿着哪个轴进行连接，默认为0，表示沿着第一个轴进行连接。对于一维数组来说，只有一个轴，所以默认值为0；  
对于二维数组来说，axis=0表示沿着行数增加的方向进行连接，axis=1表示沿着列数增加的方向进行连接。

```python
arr1 = np.random.randint(0,100,10)
arr2 = np.random.randint(0,100,10)
arr_concat = np.concatenate((arr1, arr2)) # 将arr1和arr2沿着默认的轴（即一维数组的轴）进行连接，得到一个新的数组
display(arr1)
display(arr2)
display(arr_concat)
```

```python
arr3 = np.random.randint(0,100,(3,4))
arr4 = np.random.randint(0,100,(3,4))
arr_concat2 = np.concatenate((arr3, arr4), axis=0) # 将arr3和arr4沿着行的方向进行连接，得到一个新的数组，行数增加，列数不变
display(arr3)
display(arr4)
display(arr_concat2)
```

- np.hstack()用于横向级联.  
它接受一个由数组组成的序列作为输入，并将它们沿着水平方向连接起来，返回一个新的数组.  
使用np.hstack()可以方便地将多个数组合并成一个更大的数组，特别适用于需要将多个一维数组或二维数组按列连接的情况.  
- np.vstack()用于纵向级联

```python
arr5 = np.random.randint(0,100,5)
arr6 = np.random.randint(0,100,5)
arr_concat3 = np.hstack((arr5, arr6)) # 将arr5和arr6沿着列的方向进行连接，得到一个新的数组，列数增加，行数不变
display(arr5)
display(arr6)
display(arr_concat3)
```

```python
arr7 = np.random.randint(0,100,(2,3))
arr8 = np.random.randint(0,100,(2,3))
arr_concat4 = np.vstack((arr7, arr8)) # 将arr7和arr8沿着行的方向进行连接，得到一个新的数组，行数增加，列数不变
display(arr7)
display(arr8)
display(arr_concat4)
```

### 切分
与级联类似，三个函数完成切分工作：
- np.split(): 将一个数组分割成多个子数组，返回一个列表，包含了分割后的子数组  
参数说明：  
- ary：要分割的数组  
- indices_or_sections：分割的方式，可以是一个整数，表示将数组平均分成多少份(要求切分的方向上的长度必须能被整除)
- 也可以是一个整数列表，表示分割的位置；还可以是一个整数数组，表示分割的位置  
- axis：分割的轴，默认为0，表示沿着行的方向进行分割；如果为1，表示沿着列的方向进行分割；如果为2，表示沿着深度的方向进行分割  
- 返回值：返回一个列表，包含了分割后的子数组  

- np.vsplit(arr, indices_or_sections) :功能与np.split类似，但只能沿着行的方向进行分割，即axis=0，不能指定其他轴进行分割  
- arr：要分割的数组。  
- indices_or_sections：可以是一个整数，表示要将数组平均分成多少个子数组；也可以是一个列表，表示要在数组的哪个位置进行分割。  

- np.hsplit(arr, indices_or_sections) :功能与np.split类似，但只能沿着列的方向进行分割，即axis=1，不能指定其他轴进行分割  

```python
arr = np.random.randint(0,100,(6,5)) # 创建一个6行5列的随机整数数组，范围在[0,100)之间
arr
```

```python
a1, a2 = np.split(arr,indices_or_sections=2,axis=0) # 将arr沿着行的方向分成2个子数组，得到一个包含两个数组的列表，每个数组包含原数组的一半行数，列数不变
display(a1,a2)
```

### 广播broadcast:当两个数组的形状不同时，NumPy会自动进行广播，使得它们的形状兼容，从而可以进行元素级的运算。
ndarray广播机制的两条规则  
如果
- 两个数组的后缘维度（即从末尾开始算起的维度）的轴长度相符
- 或其中的一方长度为1
则认为它们是广播兼容的。广播会在缺失和（或）长度为1的维度上进行。  
### 广播原则：  
1. 如果两个数组的维度不同，较小维度的数组会在前面添加1，直到两个数组的维度相同。
2. 如果两个数组在某个维度上的长度不同，但其中一个数组在该维度上的长度为1，那么在进行运算时，长度为1的数组会被自动扩展到与另一个数组相同的长度。
3. 如果两个数组在某个维度上的长度不同，并且其中一个数组在该维度上的长度不为1，那么它们的形状不兼容，无法进行运算。 

```python
# 例1
a = np.ones((2,3))#后缘维度为3
b = np.arange(3)#后缘维度为3
display(a, b)
display(a + b)
```

## ndarray的聚合操作
聚合操作是指对数组中的元素进行某种统计或计算，得到一个单一的结果。常见的聚合操作包括求和、平均值、最大值、最小值、标准差等。

axis参数的值可以是：
- None：对所有元素进行求和，返回一个标量。
- 0：沿着第一个轴（行）进行求和，返回一个数组，其中每个元素是对应列的和。
- 1：沿着第二个轴（列）进行求和，返回一个数组，其中每个元素是对应行的和。

```python
# 2.标准化就是将一组数据进行如下规则的转化：减去期望值再除以样本标准差，请封装一个函数来实现
def standardize(data):
    mean = np.mean(data) # 计算data数组的平均值，得到一个单一的结果
    std = np.std(data) # 计算data数组的标准差，得到一个单一的结果
    standardized_data = (data - mean) / std # 对data数组中的每个元素进行标准化操作，得到一个新的数组，元素类型与原数组相同
    return standardized_data        
```

```python
data = np.random.randint(0,100,10)
display(data)
standardized_data = standardize(data)
display(standardized_data)
```

```python
# 3.随机生成一个一维数组，比较其中是否有至少一个数据大于三倍平均值
data = np.random.randint(0,100,100)
display(data)
mean = np.mean(data) # 计算data数组的平均值，得到一个单一的结果
threshold = 3 * mean # 计算三倍平均值，得到一个单一的结果
display(threshold)
result = np.any(data > threshold) # 使用np.any()函数判断data数组中是否有至少一个元素大于threshold，如果有则返回True，否则返回False
display(result)
```

```python
# 4.生成1000行3列的一个标准正态分布的随机数数组，查询每一列是否存在至少一个数据大于该列的三倍标准差
data = np.random.normal(0, 1, (1000,3)) # 从标准正态分布中生成一个1000行3列的随机数数组，平均值为0，标准差为1
display(data)
std = np.std(data, axis=0) # 计算data数组每一列的标准差，得到一个包含每列标准差的一维数组
threshold = 3 * std # 计算每列的三倍标准差，得到一个包含每列三倍标准差的一维数组
display(threshold)
result = np.any(data > threshold, axis=0) # 使用np.any()函数判断data数组中每一列是否有至少一个元素大于对应列的threshold，如果有则返回True，否则返回False，得到一个包含每列结果的一维数组
display(result)
```
