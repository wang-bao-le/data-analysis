# 1.线图

线图通常用于展示数据随时间的变化趋势。

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建数据
x = np.linspace(0, 10, 100)
y = np.sin(x)


# 绘制线图
plt.plot(x, y)
plt.title('Line Plot')  # 设置axes标题
plt.xlabel('X-axis')  # 设置x轴标签
plt.ylabel('Y-axis')  # 设置y轴标签
plt.grid()  # 显示网格
plt.show()
```

# 2.散点图

散点图用于展示两个变量之间的关系，通常用于观察数据的分布情况和趋势。

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建数据
x = np.random.rand(50)# 生成0到1之间的50个随机数
y = np.random.rand(50)# 生成0到1之间的50个随机数

# 绘制散点图
plt.scatter(x, y,color="#22009B",alpha=1)# 设置散点颜色为蓝色
plt.title('Scatter Plot')  # 设置axes标题
plt.xlabel('X-axis')  # 设置x轴标签
plt.ylabel('Y-axis',rotation=0)  # 设置y轴标签
plt.grid()  # 显示网格
plt.show()
```

# 3.条形图

条形图用于比较不同类别的数据大小，适合展示分类数据的分布情况。

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建数据
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 15, 7, 12, 9]
# 设置字体为支持中文的字体,比如 Heiti TC（黑体），确保系统中已安装该字体
plt.rcParams['font.sans-serif'] = ['Heiti TC']  # 用于正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 绘制条形图
plt.bar(categories, values,color="#91009B",label='数据')# 设置条形颜色为紫色
plt.title('条形图示例')  # 设置axes标题
plt.xlabel('类别')  # 设置x轴标签
plt.ylabel('数值')  # 设置y轴标签
plt.legend(loc='upper left')  # 显示图例
plt.grid(axis='y')  # 显示y轴网格
plt.show()
```

# 4.直方图

直方图用于展示数据的分布情况，适合连续型数据的分析。

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建数据(正态分布的随机样本)
data = np.random.randn(1000)  # 生成100个正态分布的随机数

# 绘制直方图，bins 通常用于直方图，表示把数据分成多少个区间（箱子）。
plt.hist(data,bins=30, edgecolor='black',color="#009B00",label='数据',alpha=0.7)# 设置直方图颜色为绿色

# 设置字体为支持中文的字体,比如 Heiti TC（黑体），确保系统中已安装该字体
plt.rcParams['font.sans-serif'] = ['Heiti TC']  # 用于正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 定制图表
plt.title('直方图示例')  # 设置axes标题
plt.xlabel('数值区间')  # 设置x轴标签
plt.ylabel('频数')  # 设置y轴标签
plt.legend(loc='upper left')  # 显示图例
plt.grid()  # 显示网格
plt.show()
```

# 5.饼图

饼图用于展示各部分占整体的比例，适合展示分类数据的组成情况。

```python
import matplotlib.pyplot as plt

# 创建数据
labels = ['A', 'B', 'C', 'D'] # 设置饼图标签
sizes = [15, 30, 45, 10] # 设置饼图数据
colors = ['#FF9999','#66B3FF','#99FF99','#FFCC99']  # 设置饼图颜色
explode = [0.05, 0, 0, 0]  # 设置每个部分的偏移量
#绘制饼图
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%', # 显示百分比
    colors=colors,
    explode=explode,
    startangle=90, # 设置起始角度
    shadow=True, # 添加阴影
    wedgeprops={'edgecolor': '#FFFFFF', 'linewidth': 2}  # wedgeprops 用于设置饼图的边缘属性，这里设置边缘颜色为白色，线宽为2
    )
#定制图表
plt.title('饼图示例')  # 设置axes标题
plt.axis('equal') #  # 设置x轴和y轴比例相等，确保饼图是圆形
plt.show()
```

# 6.直方图 2d

二维直方图适合观察两个变量之间的数据分布情况，通常用于分析数据的密度和趋势。

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建数据
rng = np.random.default_rng(0)  # 设置随机数种子，确保每次生成的随机数相同
# 生成两个相关的数据变量
x = rng.normal(loc=0, scale=1, size=1000)  # 生成1000个均值为0，标准差为1的正态分布随机数
y = 0.5 * x + rng.normal(loc=0, scale=1, size=1000)  # 生成与x相关的y数据，添加一些噪声

# 创建画布
fig, ax = plt.subplots(figsize=(8, 6))  # 设置画布大小为8x6英寸
plt.hist2d(x, y, bins=30, cmap='viridis')  # 绘制二维直方图，使用viridis颜色映射，bins设置为30表示将数据分为30个区间

# 添加颜色条
plt.colorbar(label='Counts')  # 显示颜色条，并设置标签为'Counts'

# 定制图表
plt.title('二维直方图示例')  # 设置axes标题
plt.xlabel('X轴')  # 设置x轴标签
plt.ylabel('Y轴')  # 设置y轴标签

plt.tight_layout()  # 自动调整子图参数，使之填充整个图像区域
plt.show()  # 显示图表
```

# 7. 面积图

面积图是一种用于展示数据随时间变化的图表，通常用于显示时间序列数据的累积值或变化趋势。

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建数据
# 创建横轴数据
x = np.arange(1,8,1)  # 创建从1到7的整数序列
# 创建纵轴数据
y1 = np.array([3, 5, 2, 8, 7, 4, 6])  # 创建第一组纵轴数据
y2 = np.array([2, 3, 4, 5, 6, 7, 8])  # 创建第二组纵轴数据
# 创建画布和子图
fig, ax = plt.subplots(figsize=(8, 5))  # 设置画布大小

# 绘制面积图y1到x轴的面积图，设置颜色为#787EE4，透明度为0.4，并添加标签'数据1'
plt.fill_between(x, y1, color="#787EE4", alpha=0.4, label='数据1')
# 绘制面积图y1到y2，设置颜色为#FF6347，透明度为0.4，并添加标签'数据2'
plt.fill_between(x, y1, y2, color="#E40A0A", alpha=0.4, label='数据2')
# 定制图表
plt.title('面积图示例')  # 设置axes标题
plt.xlabel('X轴')  # 设置x轴标签
plt.ylabel('Y轴')  # 设置y轴标签
plt.legend(loc='upper left')  # 显示图例，位置在左上角
plt.grid(alpha=0.3)  # 显示网格

plt.tight_layout()  # 自动调整子图参数，使之填充整个图像区域
plt.show()  # 显示图表
```

# 8.热力图

热力图用于展示数据的密度和分布情况，通常用于分析数据的相关性和趋势。

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# 创建数据
rng = np.random.default_rng(0)  # 设置随机数种子，确保每次生成的随机数相同
data = rng.integers(1, 100, size=(4, 12))  # 生成一个4x12的随机整数数组，范围从1到100
# 设置行列标签
products = ['产品A', '产品B', '产品C', '产品D']  # 设置行标签
months = ['1月', '2月', '3月', '4月', '5月', '6月', '7月', '8月', '9月', '10月', '11月', '12月']  # 设置列标签

# 创建画布和子图
plt.figure(figsize=(12, 3))  # 设置画布大小为12x6
# 绘制热力图，使用seaborn库的heatmap函数，设置数据、行列标签、颜色映射为YlGnBu，显示每个单元格的数值，设置数值格式为整数，设置颜色条标签为'销售额‘
sns.heatmap(
    data, # 数据
    annot=True, # 在每个单元格中显示数值
    fmt='d', # 设置数值格式为整数
    cmap='Blues', # 设置颜色映射为YlGnBu
    square=True, # 设置单元格为正方形
    linewidths=0.5, # 设置单元格之间的线宽为0.5
    linecolor='w', # 设置单元格之间的线颜色为灰色
    xticklabels=months,  # 设置x轴标签为月份
    yticklabels=products,# 设置y轴标签为产品
    cbar_kws={'label': '销售额'}
    )
# 定制图表
plt.title('热力图示例')  # 设置axes标题
plt.xlabel('月份')  # 设置x轴标签
plt.ylabel('产品')  # 设置y轴标签
plt.tight_layout()  # 自动调整子图参数，使之填充整个图像区域
plt.show()  # 显示图表
```

# 9.三维图形

Matplotlib提供了三维绘图功能，例如三维散点图、三维曲面图等，适合展示多维数据的关系和分布情况。

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建数据
# 创建随机数种子
rng = np.random.default_rng(0)  # 设置随机数种子，确保每次生成的随机数相同
x = rng.normal(loc=0, scale=1, size=100)  # 生成100个均值为0，标准差为1的正态分布随机数
y = rng.normal(loc=0, scale=1, size=100)  # 生成100个均值为0，标准差为1的正态分布随机数
z = x + y + rng.normal(loc=0, scale=0.5, size=100)  # 生成与x和y相关的z数据，添加一些噪声

# 创建画布和三维子图
fig = plt.figure(figsize=(8, 4))  # 设置画布大小为8x6英寸
ax = fig.add_subplot(111, projection='3d')  # 创建三维子图，111表示1行1列的第1个子图，projection='3d'表示三维投影

# 绘制三维散点图，设置点的大小为50，颜色为蓝色，透明度为0.6
scatter = ax.scatter(
    x,
    y,
    zs=z,  # type: ignore[arg-type]
    c=z,# 根据z值设置颜色
    cmap ='viridis', # 设置颜色映射为viridis
    s=50,# 设置点的大小为50
    alpha=0.6# 设置点的透明度为0.6
    )

# 添加颜色条，scatter为要添加颜色条的图形对象，ax=ax表示颜色条添加到当前子图上，label='z值'表示颜色条的标签为'z值'
plt.colorbar(scatter, ax=ax, label='z值')  

# 定制图表
ax.set_title('三维散点图示例')  # 设置axes标题
ax.set_xlabel('X轴')  # 设置x轴标签
ax.set_ylabel('Y轴')  # 设置y轴标签
ax.set_zlabel('Z轴')  # 设置z轴标签

# 调整视角
ax.view_init(elev=25, azim=45)  # 设置视角，elev为仰角，azim为方位角

plt.tight_layout()  # 自动调整子图参数，使之填充整个图像区域
plt.show()  # 显示图表
```

# 10.堆叠图

堆叠图通常用于展示不同类别的数据随某个变量的变化情况，适合展示分类数据的累积值和趋势。

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建数据

#横轴数据
x = np.arange(1, 7)  # 创建从1到6的整数
#三类产品的销售额数据
y1 = np.array([10, 15, 7, 12, 9, 14])  # 产品A的销售额
y2 = np.array([8, 12, 5, 10, 7, 11])   # 产品B的销售额
y3 = np.array([6, 9, 4, 8, 5, 10])    # 产品C的销售额

# 创建画布和子图
fig, ax = plt.subplots(figsize=(8, 5))  # 设置画布大小

# 绘制堆叠图，设置颜色为#FF9999、#66B3FF、#99FF99，透明度为0.8，并添加标签'产品A'、'产品B'、'产品C'
plt.bar(x, y1, color="#FF9999", alpha=0.8, label='产品A')
plt.bar(x, y2, bottom=y1, color="#66B3FF", alpha=0.8, label='产品B')#bottom参数用于指定堆叠的起始位置，这里设置为y1表示产品B的销售额堆叠在产品A的销售额之上
plt.bar(x, y3, bottom=y1+y2, color="#99FF99", alpha=0.8, label='产品C')

# 定制图表
plt.title('堆叠图示例')  # 设置axes标题
plt.xlabel('月份')  # 设置x轴标签
plt.ylabel('销售额')  # 设置y轴标签
plt.xticks(x, labels=['1月', '2月', '3月', '4月', '5月', '6月'])  # 设置x轴刻度标签 
plt.legend()  # 显示图例
plt.grid(axis='y',alpha=0.3)  # 显示网格，透明度为0.3
plt.tight_layout()  # 自动调整子图参数，使之填充整个图像区域
plt.show()  # 显示图表
```

# 11.箱线图

箱线图用于展示数据的分布情况和异常值，适合分析数据的集中趋势和离散程度。

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建数据
rng = np.random.default_rng(0)  # 设置随机数种子，确保每次生成的随机数相同
class1 = rng.normal(loc=70, scale=10, size=100)  # 生成100个均值为70，标准差为10的正态分布随机数
class2 = rng.normal(loc=80, scale=15, size=100)  # 生成100个均值为80，标准差为15的正态分布随机数
class3 = rng.normal(loc=60, scale=5, size=100)  # 生成100个均值为60，标准差为5的正态分布随机数


# 创建画布和子图
fig, ax = plt.subplots(figsize=(6, 4))  # 设置画布大小  
# 手动加入几个异常值
class1 = np.append(class1, [30, 120])  # 在class1中添加两个异常值30和120
class2 = np.append(class2, [20, 150])  # 在class2中添加两个异常值20和150
class3 = np.append(class3, [10, 90])  # 在class3中添加两个异常
# 绘制箱线图
_=plt.boxplot(
    [class1, class2, class3],
    tick_labels=['Class 1', 'Class 2', 'Class 3'],# 设置x轴刻度标签
    patch_artist=True,  # 设置为True表示填充箱体颜色
    showmeans=True,  # 显示均值
    boxprops=dict(facecolor="#0A5FD6", color="#080101"),  # 设置箱体颜色为红色，边框颜色为黑色
    medianprops=dict(color='#FF0000'),  # 设置中位数线颜色为红色
    whiskerprops=dict(color='#080101'),  # 设置须的颜色为黑色
    capprops=dict(color='#080101'),  # 设置箱线图的顶端和底端的横线颜色为黑色
    flierprops=dict(
        marker='o', 
        markerfacecolor='#FF0000',
        markersize=5, 
        linestyle='none',
        markeredgecolor='#080101')  # 设置异常值的样式，红色圆点，大小为5
)

# 定制图表
plt.title('箱线图示例')  # 设置axes标题
plt.xlabel('班级')  # 设置x轴标签
plt.ylabel('成绩')  # 设置y轴标签

plt.grid(axis='y',alpha=0.3)  # 显示y轴网格，透明度为0.3

plt.tight_layout()  # 自动调整子图参数，使之填充整个图像区域
plt.show()  # 显示图表
```
