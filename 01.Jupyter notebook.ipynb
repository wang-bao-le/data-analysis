## 1.两种操作模式
命令模式（Command Mode）：按 Esc，对 Cell 进行管理操作（新增、删除、移动等）  
编辑模式（Edit Mode）：按 Enter，在 Cell 内部编写和修改代码或文本  

## 2.cell类型切换           
快捷键：Y&emsp;&emsp;&emsp;&emsp;切换为：Code（代码）&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;用途：编写并运行 Python 代码  
快捷键：M&emsp;&emsp;&emsp;切换为：Markdown（标记语言）&emsp;&emsp;用途：编写格式化文档、标题、说明文字

## 3.单元格操作
前提：单元格处于命令模式
### 3.1新增cell
快捷键A，在当前 Cell 上方插入新 Cell（Above）  
快捷键B，在当前 Cell 下方插入新 Cell（Below）  
### 3.2删除cell
快捷键D, D（连按两次 D）,删除当前 Cell
### 3.3复制cell
快捷键C，复制当前 Cell
### 3.4粘贴cell
快捷键V，在当前 Cell 下方粘贴 Cell
### 3.5剪切cell
快捷键X，剪切当前 Cell
### 3.6撤销cell
快捷键Z，撤销删除（恢复刚刚删除的 Cell）

## 4.运行方式
4.1快捷键：Shift + Enter，运行当前 Cell，并自动在其下方选中（新增）一个cell  
4.2快捷键：Ctrl + Enter，运行当前 Cell，并停留在当前 Cell  
4.3快捷键：Alt + Enter，运行当前 Cell，并在下方插入一个新 Cell  

## 5.魔法指令
### 魔法命令是 Jupyter 内置的特殊指令，以 %（单行）或 %%（整个 Cell）开头，用于完成计时、调试、文件操作等常用任务，无需安装任何库即可使用  
### 5.1 %run:在Jupyter中运行外部的py脚本

```python
%run main.py
```

### 5.2 %time和%%time运行计时
>%time:记录一行的运行时间  
>%%time:记录多行的运行时间

```python
def add(number):
    sum = 0
    for i in range(number):
        sum = i + number
    return sum
```

```python
%time add(1000000)
```

```python
%%time
add(1000000)
add(1000000)
```

### 5.3 %timeit和%%timeit多次运行取运行时间的平均值
>%timeit:记录一行的运行时间  
>%%timeit:记录多行的运行时间

```python
%timeit add(1000000)
```

```python
%%timeit
add(10000)
add(10000)
```

### 5.4 %who和%whos
>%who：列出当前命名空间中所有已定义的变量名.    
>%whos：比 %who 更详细，同时显示变量的类型和值

```python
%who
```

```python
%whos
```
